```python
# ğŸ“¦ Imports
import pandas as pd

# ğŸ“‚ Chemins vers les fichiers
base_path = "../data/output/"
x_train_path = base_path + "X_train.parquet"
x_test_path = base_path + "X_test.parquet"
y_train_path = base_path + "y_train.parquet"
y_test_path = base_path + "y_test.parquet"

# ğŸ“¥ Chargement des fichiers
X_train = pd.read_parquet(x_train_path)
X_test = pd.read_parquet(x_test_path)
y_train = pd.read_parquet(y_train_path)
y_test = pd.read_parquet(y_test_path)

# âœ… VÃ©rification des dimensions
assert len(X_train) == len(y_train), "âŒ Dimensions mismatch: X_train / y_train"
assert len(X_test) == len(y_test), "âŒ Dimensions mismatch: X_test / y_test"

# ğŸ”— Fusion X + y
train_df = X_train.copy()
train_df["target"] = y_train.values

test_df = X_test.copy()
test_df["target"] = y_test.values

# ğŸ”„ ConcatÃ©nation train + test
full_df = pd.concat([train_df, test_df], ignore_index=True)

# ğŸ’¾ Export CSV final
full_df.to_csv("dataset.csv", index=False)

print("âœ… Fichier 'dataset.csv' crÃ©Ã© avec succÃ¨s !")
print(f"Nombre total de lignes : {len(full_df)}")
print("ğŸ“ Emplacement : ./notebooks/dataset.csv")
```
